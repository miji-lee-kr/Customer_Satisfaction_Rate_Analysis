## 1. Data Preprocessing

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
from datetime import datetime, timedelta
import matplotlib.patches as mpatches
```

```python
items = pd.read_csv("olist_order_items_dataset.csv")
review = pd.read_csv("olist_order_reviews_dataset.csv")
product = pd.read_csv("olist_products_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
payment = pd.read_csv("olist_order_payments_dataset.csv")
translate = pd.read_csv("product_category_name_translation.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
```

```python
df = pd.merge(review, items, on= 'order_id', how = 'inner')
df= df.merge(product, on = 'product_id', how = 'inner')
df= df.merge(orders, on = 'order_id', how = 'inner')
df= df.merge(payment, on = 'order_id', how = 'inner')
df = df.merge(customers, on = 'customer_id', how = 'inner')
df.head(3)
```

```python
# dropping unnecessary columns

df = df.drop(columns = ['review_comment_title','freight_value', 'product_name_lenght','product_length_cm', 'product_height_cm', 
                        'product_width_cm', 'customer_zip_code_prefix', 'product_weight_g', 'shipping_limit_date', 'customer_city'])
```

```python
# replacing Portuguese category names with English names

df = df.join(translate.set_index('product_category_name')['product_category_name_english'], on = 'product_category_name')
df.drop(columns = 'product_category_name', inplace = True)
df.rename({'product_category_name_english': 'product_category'},axis= "columns", inplace = True)
df['product_category'].sample(3)
```

```python
# checking for datatypes - datetime & category

dtime = ['review_creation_date', 'review_answer_timestamp', 'order_purchase_timestamp', 'order_delivered_carrier_date', 'order_delivered_customer_date', 
         'order_estimated_delivery_date', 'order_approved_at']

for d in dtime:
    df[d] = df[d].astype('datetime64')

df.info()
```

```python
# checking for missing values in all columns

df['product_category'] = df['product_category'].fillna("no_data")

avg = ['product_description_lenght', 'product_photos_qty']

for i in avg:
    df[i].fillna(df[i].mean(), inplace = True)
        
# handling missing values

df["order_approved_at"].fillna(df["order_purchase_timestamp"], inplace=True)
df["order_estimated_delivery_date"].fillna(df["order_delivered_customer_date"], inplace=True)
df['order_delivered_carrier_date'].fillna(df['order_approved_at'], inplace = True)
df['order_delivered_customer_date'].fillna(df["order_estimated_delivery_date"], inplace = True)

# update reviews written in Portuguese to "review_contents", since I am not able to use the contents 

df['review_comment_message'].fillna('no_review',inplace=True)
ind = df.query('review_comment_message!= "no_review"').index 
df.loc[ind, 'review_comment_message'] = 'review_given'

df.isna().sum()
```

```python
# removing duplicates

df.drop_duplicates(subset = {'order_purchase_timestamp'}, inplace = True)
```

### Univariate Analysis

```python
# Let's first divide positive reviews with score higher than 3, and negative reviews with score lower than 3.

def division(x):
    if x<=3:
        return "0"
#     elif x>3:
#         return "2"
    else: return "1"

df['reaction'] = df['review_score'].apply(division)

palette = sns.color_palette("Paired", 2)
palette_o = sns.color_palette("YlGnBu", 10)
palette_i = sns.color_palette("Spectral",10)

outer = [palette_o[5],palette_o[4],palette_o[3],palette_o[2], palette_o[1]]
inner = [palette_i[3], palette_i[2]]
color_list = [palette[0], palette[1]]

labels = ['No Review', 'Review Given']
x1 = [df.query('review_comment_message == "no_review"').shape[0], df.query('review_comment_message == "review_given"').shape[0]]
number = df['review_score'].value_counts().sort_index(ascending=True).values
labels_2 = ['1','2', '3', '4', '5']
x2 = [number[0],number[1], number[2], number[3], number[4]]
labels_3 =['Positive','Negative']
x3 = [df['reaction'].value_counts().values[0],df['reaction'].value_counts().values[1]]

plt.figure(figsize=(9,6))
plt.pie(x1, startangle=90, labels = labels, pctdistance =0.84 , colors=color_list,autopct = '%1.0f%%',textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, radius= 0.6, labeldistance=1)
plt.pie(x2,startangle=90, labels = labels_2, pctdistance =0.73,colors=outer,textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, autopct = '%1.0f%%',radius= 0.4)
plt.pie(x3,startangle=90, labels = labels_3, pctdistance =0.4,colors=inner,textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, autopct = '%1.0f%%',radius= 0.2 )
plt.axis('equal') 
plt.legend(loc=1, bbox_to_anchor=(0.8, 0., 0.4, 1), fontsize =11)
plt.title("Overview of Reviews Given", size = 15)
plt.tight_layout();
```

- Only less than half (41%) of customers who purchased from Olist have written review. Fortunately they have provided a review score regardless of writing a review, which could be used as a good index of customer satisfaction rate.
- More than half of customers (58%) have given the highest score (5) for their shopping experience, while 11% of customers have given the lowest score (1). I have categorized review score 1 & 2 as "Negative", which account for 15% of the whole reaction, and score 4 & 5 as "Positive" which constitute 85%.

```python
# ëª¨ë¸ë§

df_model = df.drop(columns = ['review_id', 'order_id','order_item_id', 'product_id','seller_id', 'review_score', 'review_creation_date', 'review_answer_timestamp', 'order_purchase_timestamp',
                              'order_approved_at', 'customer_id', 'review_ym', 'order_date', 'order_delivered_carrier_date', 'review_comment_message',
                             'order_delivered_customer_date', 'order_estimated_delivery_date', 'delivery_time','week_num', 'customer_unique_id', 'reaction'])
df_model.isnull().sum()

# ë…ë¦½ì ì¸ ì¹¼ëŸ¼ ì‚­ì œ. ê²¹ì¹˜ëŠ”ê²Œ ì•„ë‹ˆë©´
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
    
c = df_model.columns[df_model.dtypes == 'object']
# ì¡°ê±´ì„ ë°ì´í„°í”„ë ˆìž„ì´ ì•„ë‹ˆë¼ columnsì— ê±´ë‹¤.

for i in c:
    df_model[i] = le.fit_transform(df_model[i])
    
# label encoder ë¬¸ìží˜•ìœ¼ë¡œ ë°”ê¾¸ê¸°.
# ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ ì“°ë ¤ë©´ ë¬´ì¡°ê±´ ë¬¸ìžëŠ” ìˆ«ìžë¡œ. ê¼­ catboostê°€ ì•„ë‹ˆë”ë¼ë„.
```

```python
# testì…‹ë„ ë‚˜ëˆ„ã…“ì•¼.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_model, df['reaction'], test_size = 0.2, random_state = 100)
# random_state =100 ì€ testì…‹ì´ ì‹¤í–‰í•  ë•Œë§ˆë‹¤ ë‹¬ë¼ì§€ì§€ ì•Šë„ë¡
# ë²”ì£„ìœ¨ ë°ì´í„°ì…‹ì—ì„œëŠ” x,y_validë¥¼ ë‚˜ëˆ´ìœ¼ë‚˜ ì—¬ê¸°ëŠ” ë¨¼ì € testì…‹ì„ ë½‘ì•„ë’€ë‹¤. ìºê¸€ì—ì„œ testì…‹ì„ ë”°ë¡œ ì œê³µí•˜ëŠ” ê²ƒì²˜ëŸ¼. 
# validì…‹ìœ¼ë¡œ ëª¨ë¸ì´ ìž˜ í•™ìŠµí–ˆë‚˜ ë³¸ ë‹¤ìŒ, testì…‹ìœ¼ë¡œ ë§ˆì§€ë§‰ì— 'ì´ëŸ° ê³ ê°ì€ ì´ë ‡ê²Œ í–‰ë™í•  ê²ƒì´ë‹¤' ë¼ê³  ì œì‹œí•´ì¤Œ.

# x_train: í•™ìŠµìš© ì‹œí—˜ì§€. df_model * 0.8 * 0.8
# y_train: í•™ìŠµìš© í•´ë‹µì§€

# x_valid: í•™ìŠµìš© ì‹œí—˜ì§€ 2. ì´ ëª¨ë¸ì´ í•™ìŠµì„ ìž˜ í•˜ë‚˜ ì¤‘ê°„ì¤‘ê°„ í‰ê°€í•´ì•¼
# y_valid: í•™ìŠµìš© í•´ë‹µì§€ 2

# x_test: ì‹œí—˜ìš© í•´ë‹µì§€. ì¤‘ê°„ì¤‘ê°„ ìˆ˜ì •í–ˆë˜ ëª¨ë¸ì„ í•œë²ˆë„ ëª»ë³¸ ì‹¤ì „ ë°ì´í„°ë¡œ ëŒë ¤ ì‹¤ì œë¡œ ì“°ì¼ ìˆ˜ ìžˆì„ì§€(https://lsjsj92.tistory.com/545)
# y_test: ì‹œí—˜ìš© í•´ë‹µì§€: ì´ê²Œ ìºê¸€ì— ì œì¶œí•˜ëŠ” ë‹µì§€


# ì—¬ê¸°ë¶€í„° ìˆ˜ì—…ì—ì„œ ì•ˆí•œ ë‚´ìš© - trainì„ ë‹¤ì‹œ ë‚˜ëˆ ì•¼.
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size = 0.2, random_state = 100)
x_train.shape, x_valid.shape, x_test.shape

# ì—¬ëŸ¬ ë³€ìˆ˜ì— í•œë²ˆì— í•¨ìˆ˜ ì¶”ê°€í•˜ê¸°: ctrlëˆ„ë¥´ê³  ê° ë³€ìˆ˜ì˜ ì˜¤ë¥¸ìª½ ë í´ë¦­í•˜ê³  íƒ€ì´í•‘

# ê³¼ëŒ€ì í•©. ìƒˆë¡œìš´ ë°ì´í„°. ë”°ë¡œ í‰ê°€ì…‹ ìžˆì–´ì•¼. ê°œì„ í•˜ë ¤ë©´ êµì°¨ê²€ì¦ ì‚¬ìš©
```

```python
# ì˜µì…˜ 1. regressor - ê²°ê³¼ëŠ” 1, 1.5, 2.3ì´ëŸ° ì‹ìœ¼ë¡œ ë‚˜ì˜´

from catboost import CatBoostRegressor
cbr = CatBoostRegressor(verbose = 100, depth = 10) # 10ê°œì”© í•™ìŠµí• ë•Œë§ˆë‹¤ ë­í•˜ê³  ìžˆëŠ”ì§€ ë‹¤ ë§í•´ì¤„ í•„ìš” ì—†ìŒ
cbr.fit(x_train, y_train, eval_set=(x_valid, y_valid), early_stopping_rounds=50)

# íŠ¸ë ˆì¸ì…‹ ì ìˆ˜ëŠ” ë¬´ì¡°ê±´ ì¢‹ê²Œ ë‚˜ì˜´. í‰ê°€ì…‹ evalì„ ë´ì•¼.
# 1000ë²ˆì„ ê¸°ë³¸ í•™ìŠµí•˜ì§€ë§Œ ì¤‘ê°„ì— 500-700ë²ˆì¯¤ ì ìˆ˜ê°€ ë‚˜ë¹ ì§€ëŠ” ì¼ ìžˆë‹¤. ëª¨ë¸ì´ ìŠ¤ìŠ¤ë¡œ ê·¹ë³µí•˜ëŠ” ê²½ìš°ë„ ìžˆì§€ë§Œ ì•„ì˜ˆ ë¹ ì ¸ë²„ë¦¬ë©´ = 50ë²ˆ ì—°ì† ì ìˆ˜ ê°œì„  ì—†ìœ¼ë©´ ë©ˆì¶”ë¼
```

```python
# ê²½ìš°ì˜ ìˆ˜ 1. ì˜ˆì¸¡ê¹Œì§€ í•œë²ˆì— ë‹¤ í• ê²½ìš°
# íšŒê·€ë¡œ í’€ ê²½ìš°: Kfold

from sklearn.model_selection import KFold 
kf = KFold(n_splits= 5, shuffle=True, random_state = 42)
# shuffle: ì •ë‹µ í´ëž˜ìŠ¤ê°€ í•œìª½ì— ëª°ë¦° ê²½ìš° ëŒ€ë¹„
# random_state = ê³ ì •í•´ë‘¬ì•¼. êµì°¨ê²€ì¦ ì ìˆ˜ê°€ ìš´ì´ ì¢‹ì•„ì„œ ì¢‹ì•„ì¡Œë‚˜, ì¹¼ëŸ¼ ì¶”ê°€ë¡œ ì¢‹ì•„ì¡Œë‚˜ ì•Œê¸°ìœ„í•´.

# ë°ì´í„° ìƒˆë¡œ ë½‘ê³ , ëª¨ë¸ë§í•˜ëŠ” ìˆœì„œë¡œ 5ë²ˆ ì§„í–‰.
result_cv = 0

for train_index, valid_index in kf.split(x_train): # kf.split(x_train): x_train ìžë£Œ ì•½ 7ë§Œì¤„ì—ì„œ indexë§Œ ëª¨ì•„ë…¸ì€ ê²ƒ
#     print(valid_index), print(train_index) - ì•ˆê²¹ì¹¨
    xx_train = x_train.iloc[train_index]
    xx_valid = x_train.iloc[valid_index]
    yy_train = y_train.iloc[train_index]
    yy_valid = y_train.iloc[valid_index]
    cbr_cv = CatBoostRegressor(verbose = 100)
    cbr_cv.fit(xx_train, yy_train, eval_set=(xx_valid, yy_valid), early_stopping_rounds=50)
    
    # ê°€ì¤‘ì¹˜ ì„¤ì •. í‰ê°€ì…‹ìœ¼ë¡œ ì´ë•Œê°€ ì ìˆ˜ ê°€ìž¥ ì¢‹ì€ ê°€ì¤‘ì¹˜ ì°¾ì•„ê°. ê°€ì¤‘ì¹˜ê°€ ì¢‹ì•„ì§€ê³  ìžˆë‹¤ëŠ” ëœ»-> learnìˆ«ìžê°€ ìž‘ì•„ì§€ëŠ” ê²ƒ.
    # ê° í´ë“œë§ˆë‹¤ 133ë²ˆ, 121ë²ˆ... í•™ìŠµ íšŸìˆ˜ë¡œ ìµœì ì˜ í•™ìŠµ íšŸìˆ˜ = ìµœì ì˜ í•™ìŠµ ì˜µì…˜ì„ ì°¾ì•„ê°. 
    # ì²«ë²ˆì§¸ 20%ê°€ foldì¼ ë•Œ 133ë²ˆ ë°˜ë³µí•˜ê³  2ë²ˆì§¸ 20%ê°€ foldì¼ë•ŒëŠ” 121ë²ˆ í•™ìŠµ ë°˜ë³µí•´ì„œ ì ì  ìˆ«ìžê°€ ìž‘ì•„ì§ - boosting ëª¨ë¸ì´ë‹ˆê¹Œ ì´ì „ ë‚˜ë¬´ë¥¼ ì°¸ê³ í•  ìˆ˜ ìžˆì–´ ìˆ«ìžê°€ ì ì  ìž‘ì•„ì§.
    result_cv+=cbr_cv.predict(x_test)/5 # ê° íŠ¸ë¦¬ì—ì„œ ì¸ë±ìŠ¤ë§ˆë‹¤ ì˜ˆì¸¡ë˜ëŠ” ê°’ì´ ë‹¤ë¥´ë‹ˆê¹Œ 5ë¡œ ë‚˜ëˆ ì„œ ì˜ˆì¸¡ê°’ í‰ê· ì„ êµ¬í•˜ìž
    # ì´ ì˜ˆì¸¡ê°’ í‰ê· ì„ result_cvì— ì¶”ê°€í•¨
```

```python
df_user = pd.DataFrame(df['customer_unique_id'])
df_user.columns = ['customer_unique_id']

df_max_purchase = df.groupby('customer_unique_id').order_purchase_timestamp.max().reset_index()
df_max_purchase.columns = ['customer_unique_id', 'MaxPurchaseDate']

df_max_purchase['Recency'] = (df_max_purchase['MaxPurchaseDate'].max() - df_max_purchase['MaxPurchaseDate']).dt.days
df_user = pd.merge(df_user, df_max_purchase[['customer_unique_id','Recency']], on='customer_unique_id')
df_user.head()
```

```python
df_frequency = df.groupby('customer_unique_id').order_purchase_timestamp.count().reset_index()
df_frequency.columns = ['customer_unique_id','Frequency']
df_user = pd.merge(df_user, df_frequency, on='customer_unique_id')
df_user.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.distplot(df_user['Frequency'])
plt.title("Frequency Distribution", size = 15);
```
