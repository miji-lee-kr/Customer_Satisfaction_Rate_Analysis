## 1. Data Preprocessing

````python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
from datetime import datetime, timedelta
import matplotlib.patches as mpatches
````

````python
items = pd.read_csv("olist_order_items_dataset.csv")
review = pd.read_csv("olist_order_reviews_dataset.csv")
product = pd.read_csv("olist_products_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
payment = pd.read_csv("olist_order_payments_dataset.csv")
translate = pd.read_csv("product_category_name_translation.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
````

````python
df = pd.merge(review, items, on= 'order_id', how = 'inner')
df= df.merge(product, on = 'product_id', how = 'inner')
df= df.merge(orders, on = 'order_id', how = 'inner')
df= df.merge(payment, on = 'order_id', how = 'inner')
df = df.merge(customers, on = 'customer_id', how = 'inner')
df.head(3)
````

````python
# dropping unnecessary columns

df = df.drop(columns = ['review_comment_title','freight_value', 'product_name_lenght','product_length_cm', 'product_height_cm', 
                        'product_width_cm', 'customer_zip_code_prefix', 'product_weight_g', 'shipping_limit_date', 'customer_city'])
````

````python
# replacing Portuguese category names with English names

df = df.join(translate.set_index('product_category_name')['product_category_name_english'], on = 'product_category_name')
df.drop(columns = 'product_category_name', inplace = True)
df.rename({'product_category_name_english': 'product_category'},axis= "columns", inplace = True)
df['product_category'].sample(3)
````

````python
# checking for datatypes - datetime & category

dtime = ['review_creation_date', 'review_answer_timestamp', 'order_purchase_timestamp', 'order_delivered_carrier_date', 'order_delivered_customer_date', 
         'order_estimated_delivery_date', 'order_approved_at']

for d in dtime:
    df[d] = df[d].astype('datetime64')

df.info()
````

````python
# checking for missing values in all columns

df['product_category'] = df['product_category'].fillna("no_data")

avg = ['product_description_lenght', 'product_photos_qty']

for i in avg:
    df[i].fillna(df[i].mean(), inplace = True)
        
# handling missing values

df["order_approved_at"].fillna(df["order_purchase_timestamp"], inplace=True)
df["order_estimated_delivery_date"].fillna(df["order_delivered_customer_date"], inplace=True)
df['order_delivered_carrier_date'].fillna(df['order_approved_at'], inplace = True)
df['order_delivered_customer_date'].fillna(df["order_estimated_delivery_date"], inplace = True)

# update reviews written in Portuguese to "review_contents", since I am not able to use the contents 

df['review_comment_message'].fillna('no_review',inplace=True)
ind = df.query('review_comment_message!= "no_review"').index 
df.loc[ind, 'review_comment_message'] = 'review_given'

df.isna().sum()
````

````python
# removing duplicates

df.drop_duplicates(subset = {'order_id', 'customer_id', 'order_purchase_timestamp'}, inplace = True)
````

## Data Analysis

````python
# Let's first divide positive reviews with score higher than 3, and negative reviews with score lower than 3.

def division(x):
    if x<3:
        return "Negative"
    elif x>3:
        return "Positive"
df['reaction'] = df['review_score'].apply(division)

palette = sns.color_palette("Paired", 2)
palette_o = sns.color_palette("YlGnBu", 10)
palette_i = sns.color_palette("Spectral",10)

outer = [palette_o[5],palette_o[4],palette_o[3],palette_o[2], palette_o[1]]
inner = [palette_i[3], palette_i[2]]
color_list = [palette[0], palette[1]]

labels = ['No Review', 'Review Given']
x1 = [df.query('review_comment_message == "no_review"').shape[0], df.query('review_comment_message == "review_given"').shape[0]]
number = df['review_score'].value_counts().sort_index(ascending=True).values
labels_2 = ['1','2', '3', '4', '5']
x2 = [number[0],number[1], number[2], number[3], number[4]]
labels_3 =['Positive','Negative']
x3 = [df['reaction'].value_counts().values[0],df['reaction'].value_counts().values[1]]

plt.figure(figsize=(9,6))
plt.pie(x1, startangle=90, labels = labels, pctdistance =0.84 , colors=color_list,autopct = '%1.0f%%',textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, radius= 0.6, labeldistance=1)
plt.pie(x2,startangle=90, labels = labels_2, pctdistance =0.73,colors=outer,textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, autopct = '%1.0f%%',radius= 0.4)
plt.pie(x3,startangle=90, labels = labels_3, pctdistance =0.4,colors=inner,textprops ={'fontsize':11},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" }, autopct = '%1.0f%%',radius= 0.2 )
plt.axis('equal') 
plt.legend(loc=1, bbox_to_anchor=(0.8, 0., 0.4, 1), fontsize =11)
plt.title("Overview of Reviews Given", size = 15)
plt.tight_layout();
````

- Only less than half (41%) of customers who purchased from Olist have written review. Fortunately they have provided a review score regardless of writing a review, which could be used as a good index of customer satisfaction rate.
- More than half of customers (58%) have given the highest score (5) for their shopping experience, while 11% of customers have given the lowest score (1). I have categorized review score 1 & 2 as "Negative", which account for 15% of the whole reaction, and score 4 & 5 as "Positive" which constitute 85%.

### Hypothesis 1: Certain product categories might drag the review scores down.
Since there are 72 product categories, which is too many, we are going to focus on the top 10 best selling categories, based on the number of products ordered.

````python
prod_index = df['product_category'].value_counts().sort_values(ascending=False).head(10).index
prod_cat = df.set_index('product_category').loc[prod_index,:].reset_index()
colors = sns.color_palette()[0]
orders_prod_1 = prod_cat['product_category'].value_counts().sort_values(ascending=False).index

fig = plt.figure(figsize=(10,10)) 
def func_percentage(df, col, orders):
    ax = plt.subplot(2,1,1)
    ax = sns.countplot(data = df, x= col, order = orders, color = colors)
    ax.set_title("Number of Payment Type", size =13)
    ax.set_ylabel('Count', labelpad=15)
    plt.xlabel(None)
    plt.xticks(rotation= 45)
    return ax
ax = func_percentage(prod_cat, "product_category", orders_prod_1)
ax.set_title("Top 10 Best Selling Product Categories", size =15)
total = len(prod_cat['product_category'])

def total_ratio():    
    for p in ax.patches:
        percentage = '{:.0f}%'.format(100 * p.get_height()/total)
        x = p.get_x()-0.2+ p.get_width() /2
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y))
    return ax
ax = total_ratio()
total_prod = prod_cat.groupby('product_category')['reaction'].size().reset_index()

def stacked_plot(df, total, col, orders):
    ax = plt.subplot(2,1,2)
    total = df.groupby(col)['reaction'].size().reset_index()
    positive = df[df.reaction=='Positive'].groupby(col)['reaction'].size().reset_index()
    positive['reaction'] = [i / j * 100 for i,j in zip(positive['reaction'], total['reaction'])]
    total['reaction'] = [i / j * 100 for i,j in zip(total['reaction'], total['reaction'])]
    plt.ylim(0,100)
    ax = sns.barplot(x=col,  y="reaction", data=total, color=colors, order = orders)
    ax= sns.barplot(x=col, y="reaction", data=positive, color='lightblue', order = orders)
    top_bar = mpatches.Patch(color=colors, label='Negative')
    bottom_bar = mpatches.Patch(color='lightblue', label='Positive')
    plt.legend(handles=[top_bar, bottom_bar], loc = 'upper right', bbox_to_anchor=(0.7, 0, 0.5, 1))
    plt.ylabel("Percentage (%)")
    return ax

ax = stacked_plot(prod_cat, total_prod, 'product_category', orders_prod_1)
def stacked_labels():
    patch = ax.patches
    half = int(len(patch)/2)
    for i in range(half):
        pat_1 = patch[i]
        pat_2 = patch[i+half]
        height_1 = pat_1.get_height()
        height_2 = pat_2.get_height()
        total = height_1 + height_2
        width_1 = pat_1.get_x()+pat_1.get_width()/2
        ax.text(width_1-0.2, height_2-30, "{:.0f}%".format(height_2))
        ax.text(width_1-0.2, height_1-13, "{:.0f}%".format(height_1 - height_2), color = 'w')
    return ax

ax = stacked_labels()
fig.tight_layout()
plt.title("Reaction by Category", fontsize = 15)
plt.xticks(rotation= 45);
````

- Bed_bath_table are the best selling items in the whole marketplace, followed by health_beauty and sports_leisure. 
- Do note that bed_bath_table have the highest percentage of negative reviews out of all product categories. Healthy_beauty, sports_leisure has higher satisfaction score.
- Customers of group 1(```Bed_bath_table, computers_accessories, furniture_decor, watches_gifts, telephony```) have given lower satisfaction score than group 2(`health_beauty, sports_leisure, housewares, auto, toys`). 

Now I would like to drill down to find out what's caused the difference between the two groups.

#### Hypothesis 1-1. Average delivery period differs by product category

````python
# ì™œ bedê°€ ë‚®ê³  toyê°€ ë†’ì€ê°€? ê²°ì • ìš”ì¸ì€ ë¬´ì—‡ì¼ê¹Œ? ì¹´í…Œê³ ë¦¬ë³„ review scoreë¥¼ - ë°°ì†¡ ê¸°ê°„(days_til_arr), ê°€ê²©, ì§€ì—­

# ê° ì§€ì—­ì˜ (ì „ì²´ì¹´í…Œê³ ë¦¬) í‰ê·  ì ìˆ˜ & 
````

````python
prod_cat['days_til_arr'] = (prod_cat['order_delivered_customer_date'] - prod_cat['order_approved_at']).dt.days
cat_lists = ['bed_bath_table', 'telephony','furniture_decor','computers_accessories','watches_gifts','auto','housewares',
'health_beauty', 'toys','sports_leisure']
cat_var = pd.api.types.CategoricalDtype(ordered = True, categories = cat_lists)
prod_cat['product_category'] = prod_cat['product_category'].astype(cat_var)
prod_cat['price_range'] = pd.cut(prod_cat['price'], [prod_cat['price'].min(),100,300,500,700,1000,2000,3000,5000,prod_cat['price'].max()], 
                                 labels = ['-$100','-$300', '-$500','-$700','-$1000','-$2000','-$3000', '-$5000','$5000+'], 
       include_lowest = False, ordered = True, right = True)

prod_cat_1 = prod_cat.groupby(['product_category']).mean().days_til_arr.reset_index(name = 'mean_days')
prod_cat_2 = prod_cat.groupby(['product_category']).mean().price.reset_index(name = 'mean_price')
# prod_cat_2 = prod_cat.pivot_table(values = 'review_score', index = 'product_category', columns = 'price_range').fillna(0)
prod_cat_3 = prod_cat.groupby(['product_category', 'price_range']).mean().review_score.reset_index(name = 'mean_score')
````

````python
# ê°€ì„¤ 1: ê°€ê²©ì´ ë†’ì€ ì¹´í…Œê³ ë¦¬ì— ë” í‰ê°€ì— ê´€ëŒ€í•´ì§„ë‹¤. bedê°€ ë” ë‚®ì€ê±´ ê°€ê²©ì´ ë‚®ê¸° ë•Œë¬¸ì´ë‹¤.
# bedì¹´í…Œê³ ë¦¬ê°€ healthyë³´ë‹¤ ë‚®ì€ê±´ ì•Œê³ ìžˆë‹¤. ê·¸ëŸ¬ë©´ í•œ ì¹´í…Œê³ ë¦¬ ë‚´ì—ì„œëŠ” ì–´ë–¨ê¹Œ? ê°€ê²©ì´ ë†’ì„ìˆ˜ë¡ í‰ì ì´ ë†’ì•„ì§€ë‚˜?
 
# bed ì¹´í…Œê³ ë¦¬ì˜ ê°€ê²©ì´ ë‚®ì•„ì„œ ë” ë°•í•˜ê³  healthyì˜ ê°€ê²©ì´ ë†’ì•„ì„œ ë” í‰ê°€ê°€ ì¢‹ë‹¤. ê·¸ëŸ¼ bedëŠ” ë™ì¼ ê°€ê²©ëŒ€ì—ì„œ watchë³´ë‹¤ í‰ê°€ê°€ ë°•í•´ì•¼í•œë‹¤.


plt.figure(figsize = (10,5))
ax = sns.barplot(x ='product_category', y = 'mean_days', data = prod_cat_1, color = colors)
ax.set_title("Avg. Delivery Period by Each Category (days)", size = 15)
plt.xticks(rotation = 45)
plt.axvline(x = 4.5, color = 'r', linestyle = '--')
avg = prod_cat_1.mean_days.mean()
plt.axhline(y=avg, color='r', linestyle='-')
plt.text(2.6,11.3, 'All Category Average Score: {}'.format(round(avg,2)), fontsize=10, va='center', ha='center', backgroundcolor='w')

def avg_score():
    for p in ax.patches:                 
        ax.annotate(np.round(p.get_height(),decimals=1), (p.get_x()+p.get_width()/2., p.get_height()-0.4),
                    ha='center',va='center',xytext=(0, 10),textcoords='offset points')
    return ax
ax = avg_score()
````

- We can easily see that the categories in group 1 tend to take longer to deliver than group 2. 
- In order to lift up the mean score we will have to find out the root cause of long delivery period and get rid of it.

````python
outlier_top10 = df_late_prod.groupby(['product_category', 'late_outliers']).mean().days_til_arr.dropna().reset_index()
outlier_top10['product_category'] = outlier_top10['product_category'].astype(cat_var)

plt.figure(figsize = (10,6))
color_list ={"Late": sns.color_palette("RdBu",10)[8], "Not Late":sns.color_palette("RdBu",10)[6]}
ax = sns.barplot(data = outlier_top10, x = 'product_category', y ='days_til_arr', hue = 'late_outliers',palette = color_list)           
plt.xticks(rotation = 45)

def increase_rate():
    patch = ax.patches
    half = int(len(patch)/2)
    for i in range(half):
        pat_1 = patch[i]
        pat_2 = patch[i+half]
        height_1 = pat_1.get_height()
        height_2 = pat_2.get_height()
        total = height_1 + height_2
        width_1 = pat_1.get_x()+pat_1.get_width()/2
        width_2 = pat_2.get_x()+pat_2.get_width()/2
        ax.text(width_1-0.2, height_1+0.2, "{:.0f}%".format(100*(height_1-height_2)/height_2))
    return ax

ax = increase_rate()
ax.set_title('Avg. Delivery Days & Increase Rate by Category', fontsize=15)
ax.legend(loc = 'center right',title = 'Delivery Promptness')
plt.axvline(x = 4.5, color = 'r', linestyle = '--');
````
